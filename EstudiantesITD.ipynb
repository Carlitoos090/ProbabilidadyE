# Ejercicio 2
Una revision de los pesos de 500 estudiantes del ITD, evidencio que la meia es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente.
    

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# parametro de la distribucion
media = 70 #Media
desv_std = 4 #Desviacion estandar
Total_Estudiantes = 500 #Total de estudiantes

# Variable para calculo de la probabilidad de que un estudiante que pese de 75 kg
Peso_Limite = 79
# Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la funcion de distribucion acumulada
#(CDF) de la distribucion normal.
Prob_menos_75kg = norm.cdf(Peso_Limite, media, desv_std)
Estudiantes_mayor_75kg = Total_Estudiantes * Prob_menos_75kg
print("Estudiantes que pesan menos de 75 kg:", Prob_menos_75kg)
print("Estudiantes que pesan mas de 75 kg:", Estudiantes_mayor_75kg)
x = np.linspace(media - 4* desv_std, media + 4* desv_std, 1000)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2) # Corrected the typo here
plt.fill_between(x, y, where=(x<=Peso_Limite), facecolor='skyblue', alpha=0.5)
plt.title( 'Probabilidad de que un estudiante pese menos de 75 kg. \n  $\mu=70$, $\sigma=4$, x=500   ')
plt.xlabel('Peso (kg)')
plt.ylabel('Dencidad de Probabilidad')
plt.grid(True)
plt.show()
```
